```python
import pandas as pd
log_df = pd.read_csv('~/Documents/Git/skillfactory/Unit3/dataSource/log.csv', header=None)
log_df.columns = ['user_id', 'time', 'bet', 'win']
display(log_df.head(2))
display(log_df.info())

users_df = pd.read_csv('~/Documents/Git/skillfactory/Unit3/dataSource/users.csv', sep='\t', encoding='KOI8-R')
users_df.columns = ['user_id', 'email', 'geo']

def clean_time(x):
    if type(x) is str:
        return x.split('[')[1]
log_df['time']=log_df['time'].apply(clean_time)

log_df = log_df.dropna(subset=['time'], axis=0)
display(log_df.head(2))
display(log_df.info())
log_df['time2'] = log_df['time'].apply(lambda x:x[:1])

log_df.time = pd.to_datetime(log_df.time, errors='coerce')
display(users_df.head(2))
```
