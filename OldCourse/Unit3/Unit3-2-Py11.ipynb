```python
import pandas as pd
log_df = pd.read_csv('~/Documents/Git/skillfactory/Unit3/dataSource/log.csv', header=None)
log_df.columns = ['user_id', 'time', 'bet', 'win']
log_df['bet'] = log_df.bet.fillna(0)

display(log_df.bet.value_counts())
display(log_df.head(5))
```
