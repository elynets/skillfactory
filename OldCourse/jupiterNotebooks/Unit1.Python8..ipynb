```python
#8.11.3
import pandas as pd

def profession_code(s):
    if s=="Рабочий":
        return 0
    elif s=="Менеджер":
        return 1
    else:
        return 2
    
    
sample = pd.read_csv('~/Documents/Python/Unit1Data/sample.xls')
sample2 = sample
sample2["Profession"] = sample2["Profession"].apply(profession_code)

#display(sample)
display(sample2)
```

```python
import pandas as pd

sample = pd.read_csv('~/Documents/Python/Unit1Data/sample.xls')

def age_category(age):
    if(age<23):
        return "молодой"
    elif(23<age<35):
        return "средний"
    else:
        return "зрелый"
    
sample['Age_category']=sample.Age.apply(age_category)
display(sample)
```
