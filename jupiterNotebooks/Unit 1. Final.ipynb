```python

data2 = data['genres'].str.split('|')
data2 = data2.explode('genres')
print(data2.value_counts())

display(data2)
```

```python
data2 = data[data['profit']>0].genres.str.split('|')
data2 = data2.explode('genres')
print(data2.value_counts())

data2 = data[['profit', 'genres', 'imdb_id']].copy()
data2['genres'] = data2[data2['profit']>0].genres.str.split('|')
print(data2.explode('genres').groupby(['genres']).imdb_id.count().sort_values(ascending = False))
```

```python
data2 = data.copy()

df = data2.groupby(['director']).revenue.sum().sort_values(ascending = False)
display(df)
```

```python
data2 = data[['director', 'genres', 'imdb_id']].copy()
data2 = data2[data2.genres.str.contains('Action', na = False)]
data2['director'] = data2.director.str.split('|')
print(data2.explode('director').groupby(['director']).imdb_id.count().sort_values(ascending = False))

```

```python

data2 = data[['cast', 'revenue', 'release_year', 'genres', 'imdb_id']].copy()
data2 = data2[data2['release_year']==2012]
data2['cast'] = data2.cast.str.split('|')
print(data2.explode('cast').groupby(['cast']).revenue.sum().sort_values(ascending = False))
```

```python
data2 = data[['original_title', 'profit', 'production_companies', 'imdb_id']].copy()
data2 = data2[data2.production_companies.str.contains('Paramount Pictures', na = False)]

print(data2[data2.profit==data2.profit.min()].original_title)
```
